### 1. Importing pacakge and load a csv
in this example, we will show you how to read the [POPS data](https://data.cityofnewyork.us/City-Government/Privately-Owned-Public-Spaces-POPS-/rvih-nhyn/data) on NYC Open data directly into a pandas dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://data.cityofnewyork.us/api/views/rvih-nhyn/rows.csv')
```

```python
# save the csv
df.to_csv('dcp_pops.csv')
```

```python
df.head()
```

##### let's list all the columns

```python
df.columns
```

##### Take a look of the shape of the dataframe: (row, columns)

```python
df.shape
```

#### If you want to select a subset of the table:

```python
df.loc[:,['Building_Name', 'Building_Location', 'Year_Completed']].head()
```

#### We can also __aggregate__

```python
df.groupby('Year_Completed').agg({'Year_Completed':'count'})
```

```python
df.groupby('Year_Completed').agg({'Year_Completed':'count'}).plot(kind='bar', 
                                                                  figsize=(15, 8), 
                                                                  title='count of pops by year completed')
```

#### __Filter__ 

```python
df[df.Borough == 'Queens'].groupby('Year_Completed').agg({'Year_Completed':'count'}).plot(kind='bar', 
                                                                  figsize=(15, 8), 
                                                                  title='count of pops by year completed in Queens')
```

### What if we want to put these points on maps?
you can do that too! but first of all you need to install cartoframes, you can easily install any python package using pip (the python package manager)

```python
! pip install cartoframes
```

```python
import cartoframes
from cartoframes import Credentials
import os

USERNAME = 'USERNAME' 
APIKEY = 'APIKEY'
creds = Credentials(username=USERNAME, 
                    key=APIKEY)
cc = cartoframes.CartoContext(creds=creds)
```

```python
cc.write(df, 'dcp_pops', 
         lnglat=('Longitude', 'Latitude'),
         overwrite=True)
```

```python
df = cc.read('dcp_pops')
```

```python
from cartoframes import Layer
pops_map=cc.map(layers=Layer('dcp_pops'),
       interactive=True)
pops_map
```
